````python
def calculate_variance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    
    sum_squared_diff = 0  # Sum of squared differences from the mean
    for n in nl:
        squared_diff = (n - mean) ** 2
        sum_squared_diff += squared_diff
    variance = sum_squared_diff / len(nl)
    
    return variance
````
